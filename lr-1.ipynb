```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
data = pd.DataFrame(data=wine.data, columns=wine.feature_names)
data['target'] = wine.target
data.head()
```

```python
data.info()
```

```python
_, axes = plt.subplots(data.columns.__len__(), figsize=(10, 50))
i = 0
for col in data.columns:
    sns.histplot(data[col], ax=axes[i])
    i+=1
plt.subplots_adjust(hspace=0.4, wspace=0.3)
plt.show()
```

```python
data.corr()
```

```python
_, axes = plt.subplots(1, 2, figsize=(20, 4))
sns.heatmap(data.drop('target', axis=1).corr(), annot=True, vmin=-1, vmax=1, cmap='seismic', linewidth=1, ax=axes[0])
sns.heatmap(pd.DataFrame(data.corr()['target'].sort_values(ascending=False)[1:]),
            annot=True, vmin=-1, vmax=1, cmap='seismic', linewidth=1, ax=axes[1])
plt.subplots_adjust(wspace=1)
plt.show()
```
